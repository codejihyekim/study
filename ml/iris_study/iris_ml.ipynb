```python
s = open('./data/iris1.csv').readline()
labels = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
iris = np.loadtxt('./data/iris1.csv', delimiter=',', skiprows=1)
```

```python
print(iris)
#print(iris.shape, iris[:5])
```

```python
x = iris[:,:3]
y = iris[:, 3]
print(x)
print(x.shape, y.shape)
print(x[:5], y)
```

```python
corr = np.corrcoef(x.T)
print(corr)
```

```python
plt.imshow(corr, interpolation='none', vmin=-1, vmax=1, cmap='spring')
plt.xticks(range(3), header, rotation=90)
plt.yticks(range(3), header)
plt.colorbar()
```

```python
plt.boxplot(x)
plt.xticks(range(1, 4), header, rotation=90)
```

```python
plt.figure(figsize=[12,8])
for col in range(3):
    plt.subplot(2,2,col+1)
    plt.scatter(x[:, col], y + np.random.normal(0,0.03,size=len(y)), c=y, s=30, alpha=0.3)
    plt.yticks([0,1,2], ['Setosa', 'Versicolor', 'Vifginica'], rotation=90)
    plt.title(header[col], fontsize=15)
```
